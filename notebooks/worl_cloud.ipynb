```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

text1 = """ AccountingNumbers
EconomicVariables
EventStudy
DailyValuation
DeliveryQuantity 
DispositionEffect
Distribution
MutualFund
NewShare
Seasonality
VolatilitySmile
"""

# Custom colormap with brown, black, and orange
colors = ["#E18C09","#8B4513", "#000000", "#FFA500", "#B4442C", '#E1B609' , "#E11909"]  # Brown, Black, Orange

# # Generate a mask using a shape (optional)
# mask_path = "path/to/your/mask/image.png"
# mask = np.array(Image.open(mask_path)) if mask_path else None

# Create the WordCloud object with custom colors and colormap
wordcloud = WordCloud(max_font_size=50, max_words=20, background_color="white", colormap=plt.cm.colors.ListedColormap(colors),
                      prefer_horizontal=0.5).generate(text1)

# Display the generated word cloud using matplotlib
plt.figure(figsize=(16, 8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.savefig('/home/ujjwal/Documents/text1.svg', format='svg', dpi=300)
plt.show()
```
